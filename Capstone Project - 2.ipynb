```python
def soloinsulin(df):
    lst=[]
    for i in range(len(df)):
        if df.iloc[i,47] >=2:
            lst.append(0)
        else:
            if df.iloc[i,38] ==1:
                lst.append(1)
            else:
                lst.append(0)
    return lst
            
```
