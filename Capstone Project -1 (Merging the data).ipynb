```python
z=pd.merge(diagnostic,ad,left_index=True,right_index=True)
```

```python
z.drop('patient_nbr_y',axis=1,inplace=True)
z.rename({'patient_nbr_x':'patient_nbr'},axis=1,inplace=True)
z.head(10)
```
